**Q1. What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.**

ANS.

### Anomaly Detection
Anomaly detection, also known as outlier detection, is the process of identifying rare items, events, or observations that deviate significantly from the majority of the data. These anomalies often indicate some kind of problem or unusual event, such as a credit card fraud, a malfunctioning machine, or a rare disease.

### Types of Anomalies

1.  **Point Anomalies (Global Outliers):**
    *   **Definition:** A single data instance that is significantly different from the rest of the data. It's an individual data point that lies outside the overall distribution of the dataset.
    *   **Example:**
        *   **Credit Card Fraud:** A single transaction of \$10,000 made from a credit card that typically has transactions under \$100. This single, unusually large transaction, regardless of context, is a point anomaly.
        *   **Sensor Reading:** A temperature sensor in a server room suddenly reports 150°C, while all other readings are consistently between 20-25°C. This single, extreme reading is a point anomaly.

2.  **Contextual Anomalies (Conditional Anomalies):**
    *   **Definition:** A data instance that is anomalous in a specific context but not when considered out of context. The abnormality is dependent on the surrounding data or conditions.
    *   **Example:**
        *   **Website Traffic:** A website receiving 10,000 visitors per day is normal during a weekday but becomes an anomaly if it happens at 3 AM on a Sunday, when traffic is typically very low. The number of visitors itself isn't anomalous, but its occurrence at that specific time is.
        *   **Stock Price:** A 10% drop in a stock's price might be normal during a market crash (context). However, a 10% drop on a day when the rest of the market is stable and rising would be a contextual anomaly.

3.  **Collective Anomalies:**
    *   **Definition:** A collection of related data instances that, as a group, are anomalous, even though each individual instance within the collection might not be anomalous by itself. The anomaly is in the collective behavior.
    *   **Example:**
        *   **Network Intrusion:** A computer network experiences a sustained period of unusually high outbound data transfer, even if no single data packet or connection appears suspicious individually. The collective pattern of many small, legitimate-looking transfers, when viewed together, indicates a data exfiltration attempt (a collective anomaly).
        *   **Human Activity:** A person who usually wakes up at 7 AM, goes to work, and comes back at 6 PM. If, for several consecutive hours, their phone shows GPS data indicating they are at an unknown location, then at an airport, and then in another country, each individual GPS point might not be an anomaly, but the sequence of locations collectively represents an anomalous travel pattern.

### Key Components of a Time Series

A time series is a sequence of data points indexed (or listed or graphed) in time order. Most commonly, a time series is a sequence taken at successive equally spaced points in time. Understanding the components of a time series is crucial for accurate forecasting and analysis. The four main components are:

1.  **Trend:**
    *   **Definition:** The long-term increase or decrease in the data over time. It represents the underlying direction of the series, ignoring short-term fluctuations.
    *   **Example:** The global average temperature has shown an upward trend over the last century, indicating global warming. Even with year-to-year variations, the overall pattern is a clear increase.

2.  **Seasonality:**
    *   **Definition:** Patterns that repeat over a fixed period, such as a day, week, month, or year. These patterns are often driven by calendar-related events or natural cycles.
    *   **Example:** Retail sales often show a seasonal pattern, with peaks during holiday seasons (e.g., Christmas, Black Friday) and troughs in quieter months (e.g., January, February). This pattern repeats annually.

3.  **Cyclicality:**
    *   **Definition:** Fluctuations in the data that are not of a fixed period (unlike seasonality) and usually last longer than a year. These patterns are often associated with economic or business cycles.
    *   **Example:** The housing market experiences cyclical patterns. There might be periods of several years with increasing housing prices and demand, followed by periods of decreasing prices and reduced demand, which do not occur at fixed intervals but rather in waves.

4.  **Irregularity (Noise/Residuals):**
    *   **Definition:** The random or unpredictable fluctuations in the data that cannot be explained by trend, seasonality, or cyclicality. This is what's left after all other components are removed.
    *   **Example:** A sudden, unexpected surge in electricity consumption on a particular day due to an unforeseen extreme heatwave, or a sudden dip in stock prices due to an unexpected geopolitical event, are examples of irregular components. These are one-off events that don't follow a predictable pattern.

**Q4.Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?**

### Stationarity in Time Series

A **stationary time series** is one whose statistical properties (such as mean, variance, and autocorrelation) do not change over time. This means it has:

*   **Constant Mean:** The average value of the series remains constant over time.
*   **Constant Variance:** The variability of the series around its mean remains constant over time.
*   **Constant Autocovariance/Autocorrelation:** The correlation between the series and its lagged versions remains constant over time, regardless of the time lag or the point in time.

Stationarity is a crucial assumption for many time series forecasting models (e.g., ARIMA models), as non-stationary series can lead to spurious regressions and unreliable forecasts.

### How to Test for Stationarity

There are both visual and statistical methods to test for stationarity:

1.  **Visual Inspection:**
    *   **Plotting the Time Series:** Look for a constant mean (series fluctuating around a horizontal line), constant variance (fluctuations don't increase or decrease with time), and no obvious trends or seasonal patterns.
    *   **Rolling Statistics:** Plot the rolling mean and rolling standard deviation over time. If these statistics remain relatively constant, the series is likely stationary.
    *   **Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) Plots:** For a stationary series, the ACF plot will drop to zero relatively quickly (often exponentially or geometrically). For non-stationary series, the ACF tends to decay slowly.

2.  **Statistical Tests:**
    *   **Augmented Dickey-Fuller (ADF) Test:** This is a popular statistical test for stationarity. The null hypothesis (H0) is that the time series is non-stationary (it has a unit root). The alternative hypothesis (Ha) is that the time series is stationary.
        *   **Interpretation:** If the p-value is less than a chosen significance level (e.g., 0.05), you reject the null hypothesis and conclude that the series is stationary.
    *   **Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test:** This test is an alternative to ADF. The null hypothesis (H0) is that the time series is trend-stationary. The alternative hypothesis (Ha) is that the series is non-stationary.
        *   **Interpretation:** If the p-value is less than the significance level, you reject the null hypothesis and conclude that the series is non-stationary.

### How to Transform a Non-Stationary Series into a Stationary One

If a time series is found to be non-stationary, several techniques can be applied to transform it into a stationary one:

1.  **Differencing:**
    *   **Description:** This is the most common technique. It involves computing the difference between consecutive observations. First-order differencing means `Y_t = X_t - X_{t-1}`. If the series still isn't stationary, you can apply second-order differencing, and so on.
    *   **Use Case:** Effective for removing trends and some forms of seasonality. If the trend is linear, first-order differencing often suffices. For seasonal patterns, seasonal differencing (e.g., `Y_t = X_t - X_{t-season_length}`) can be used.

2.  **Log Transformation:**
    *   **Description:** Applying a natural logarithm (or any other base logarithm) to the series. `Y_t = log(X_t)`.
    *   **Use Case:** Useful for stabilizing variance (when the variance increases with the mean) and can help in converting an exponential trend into a linear one.

3.  **Square Root or Cube Root Transformation:**
    *   **Description:** Similar to log transformation, these can also help stabilize variance, especially when dealing with count data.
    *   **Use Case:** Applied when log transformation is too strong or the data contains zeros, making log transformation impossible.

4.  **Decomposition:**
    *   **Description:** Separating the time series into its constituent components: trend, seasonality, and residuals (irregularity). The residuals component is often stationary after removing the trend and seasonal components.
    *   **Use Case:** Useful when both trend and seasonality are present and need to be explicitly modeled or removed.

5.  **Detrending:**
    *   **Description:** Removing the trend component from the time series. This can be done by regressing the series on time or by using moving averages.
    *   **Use Case:** When only a trend component is making the series non-stationary, and other statistical properties are already stable.

6.  **Seasonal Adjustment:**
    *   **Description:** Removing the seasonal component from the time series, often done after detrending or in conjunction with differencing.
    *   **Use Case:** When seasonality is the primary cause of non-stationarity.

Combining these techniques (e.g., log transformation followed by differencing) is often necessary to achieve stationarity.

### Differentiation between AR, MA, ARIMA, SARIMA, and SARIMAX Models

These models are fundamental in time series analysis and forecasting, each building upon the previous one to handle increasing complexity in data patterns.

| Model   | Full Name                           | Structure                                                                                                  | Key Parameters                                        | Application/Use Cases                                                                                                                                                                                                                         | Limitations                                                                                                                                                                                                         |
| :------ | :---------------------------------- | :--------------------------------------------------------------------------------------------------------- | :---------------------------------------------------- | :-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- | :------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------ |
| **AR(p)** | AutoRegressive Model                | Forecasts current value based on a linear combination of its past *p* values (lags). Only uses past observations. `Y_t = c + φ_1 Y_{t-1} + ... + φ_p Y_{t-p} + ε_t` | `p` (order of the AR component)                       | - Modeling time series with inertia (e.g., stock prices, temperature).
- When current observations are strongly correlated with previous ones.                                                                                                                                            | - Assumes stationarity (constant mean, variance, autocovariance).
- Cannot capture sudden shocks or noise effectively.                                                                                                                                    |
| **MA(q)** | Moving Average Model                | Forecasts current value based on a linear combination of past *q* forecast errors (shocks/noise terms). `Y_t = c + ε_t + θ_1 ε_{t-1} + ... + θ_q ε_{t-q}` | `q` (order of the MA component)                       | - Modeling time series with short-term memory of past shocks (e.g., unexpected demand fluctuations).
- When current observations are influenced by random shocks in the near past.                                                                                                                                        | - Assumes stationarity.
- Cannot capture trends or strong autoregressive patterns.                                                                                                                                                           |
| **ARIMA(p, d, q)** | AutoRegressive Integrated Moving Average | Combines AR and MA components after differencing the series *d* times to achieve stationarity. `(1 - Σφ_i L^i) (1-L)^d Y_t = c + (1 + Σθ_j L^j) ε_t` | `p` (AR order), `d` (differencing order), `q` (MA order) | - Widely used for forecasting non-seasonal, univariate time series with trends (e.g., economic indicators, sales data without strong seasonality). | - Does not handle seasonality directly.
- Assumes linearity and constant variance after differencing.                                                                                                                                                         |
| **SARIMA(p, d, q)(P, D, Q)s** | Seasonal AutoRegressive Integrated Moving Average | Extends ARIMA to include seasonal components. It has both non-seasonal (p, d, q) and seasonal (P, D, Q)s orders. `(1 - Σφ_i L^i)(1 - ΣΦ_i L^s) (1-L)^d (1-L^s)^D Y_t = c + (1 + Σθ_j L^j)(1 + ΣΘ_j L^s) ε_t` | `p, d, q` (non-seasonal orders), `P, D, Q` (seasonal orders), `s` (seasonal period) | - Forecasting time series with clear and recurring seasonal patterns (e.g., monthly retail sales, quarterly electricity consumption). | - Can be complex to tune due to many parameters.
- Still assumes linearity and constant variance after differencing.                                                                                                                                                  |
| **SARIMAX(p, d, q)(P, D, Q)s** | Seasonal AutoRegressive Integrated Moving Average with Exogenous Regressors | Extends SARIMA by incorporating exogenous (external) variables. Allows for the influence of other predictor variables on the time series. `SARIMA component + β_1 X_{1,t} + ... + β_k X_{k,t}` | Same as SARIMA, plus coefficients for exogenous variables | - Forecasting time series where external factors significantly impact the series (e.g., predicting sales using advertising spend, weather, holidays).
- More comprehensive modeling when causal relationships are known. | - Requires exogenous variables to be stationary and available for future periods (for forecasting).
- Increased complexity in model building and interpretation due to additional variables.                                                                                                                               |

**Q7. Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Try to load a public NYC Taxi Fare dataset
try:
    # This URL is for a small sample of NYC taxi data, often used in examples
    df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/nyc-taxi-data.csv')


    # If the above link doesn't work or has different column names, adjust or create dummy data.
    if 'fare_amount' not in df.columns or 'trip_distance' not in df.columns:
        print("Dataset columns 'fare_amount' or 'trip_distance' not found. Will use a dummy dataset.")
        raise ValueError("Missing expected columns")

    data = df[['fare_amount', 'trip_distance']].dropna()
    print("Loaded NYC Taxi Fare data from URL.")

except Exception as e:
    print(f"Could not load NYC Taxi Fare data from URL: {e}")
    print("Generating a dummy numerical dataset for demonstration.")

    # Generate a dummy dataset resembling taxi fare data
    np.random.seed(42)
    n_samples = 1000

    # Normal data points
    fare_amount = np.random.normal(loc=10, scale=3, size=n_samples)
    trip_distance = np.random.normal(loc=3, scale=1, size=n_samples)

    # Introduce some anomalies:
    # High fare for short distance
    fare_anomaly = np.random.normal(loc=50, scale=10, size=20)
    distance_anomaly_short = np.random.normal(loc=0.5, scale=0.1, size=20)

    # Low fare for long distance
    fare_anomaly_low = np.random.normal(loc=5, scale=1, size=20)
    distance_anomaly_long = np.random.normal(loc=20, scale=5, size=20)

    # Combine
    fare_amount = np.concatenate([fare_amount, fare_anomaly, fare_anomaly_low])
    trip_distance = np.concatenate([trip_distance, distance_anomaly_short, distance_anomaly_long])

    data = pd.DataFrame({
        'fare_amount': fare_amount,
        'trip_distance': trip_distance
    })

    # Ensure non-negative values
    data['fare_amount'] = data['fare_amount'].apply(lambda x: max(0, x))
    data['trip_distance'] = data['trip_distance'].apply(lambda x: max(0.1, x))
    print("Dummy dataset created.")

# Display the first few rows and basic stats
display(data.head())
display(data.describe())
```

```python

model = IsolationForest(contamination=0.05, random_state=42)
model.fit(data)

# Predict anomalies (1 for inliers, -1 for outliers)
data['anomaly'] = model.predict(data)

print("Isolation Forest applied. Number of anomalies detected:")
print(data['anomaly'].value_counts())
```

```python
# Visualize the anomalies on a 2D scatter plot
plt.figure(figsize=(12, 8))

# Plot normal points
plt.scatter(data[data['anomaly'] == 1]['trip_distance'],
            data[data['anomaly'] == 1]['fare_amount'],
            c='blue', label='Normal Points', alpha=0.6, s=20)

# Plot anomalous points
plt.scatter(data[data['anomaly'] == -1]['trip_distance'],
            data[data['anomaly'] == -1]['fare_amount'],
            c='red', label='Anomalies', alpha=0.8, s=50, marker='X')

plt.title('Anomaly Detection using Isolation Forest (NYC Taxi Fare Data)')
plt.xlabel('Trip Distance')
plt.ylabel('Fare Amount')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

**Q9.Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Ensure 'data' DataFrame is available. Re-create if necessary for standalone execution.
if 'data' not in locals():
    print("'data' DataFrame not found. Re-generating a dummy numerical dataset for demonstration.")
    np.random.seed(42)
    n_samples = 1000

    fare_amount = np.random.normal(loc=10, scale=3, size=n_samples)
    trip_distance = np.random.normal(loc=3, scale=1, size=n_samples)

    fare_anomaly = np.random.normal(loc=50, scale=10, size=20)
    distance_anomaly_short = np.random.normal(loc=0.5, scale=0.1, size=20)

    fare_anomaly_low = np.random.normal(loc=5, scale=1, size=20)
    distance_anomaly_long = np.random.normal(loc=20, scale=5, size=20)

    fare_amount = np.concatenate([fare_amount, fare_anomaly, fare_anomaly_low])
    trip_distance = np.concatenate([trip_distance, distance_anomaly_short, distance_anomaly_long])

    data = pd.DataFrame({
        'fare_amount': fare_amount,
        'trip_distance': trip_distance
    })
    data['fare_amount'] = data['fare_amount'].apply(lambda x: max(0, x))
    data['trip_distance'] = data['trip_distance'].apply(lambda x: max(0.1, x))
    print("Dummy dataset created.")

# Initialize and train the Local Outlier Factor model
# n_neighbors: Number of neighbors to consider for the local density calculation.
# contamination: The proportion of outliers in the data set.
# novelty=False for outlier detection (fit_predict) and novelty=True for novelty detection (fit, then predict).
# Here, we use novelty=False to directly detect outliers in the training data.
lof_model = LocalOutlierFactor(n_neighbors=20, contamination=0.05, novelty=False)

# Fit the model and predict anomalies (-1 for outliers, 1 for inliers)
# Note: LOF's fit_predict directly returns the anomaly labels.
data['lof_anomaly'] = lof_model.fit_predict(data[['fare_amount', 'trip_distance']])

print("Local Outlier Factor applied. Number of anomalies detected:")
print(data['lof_anomaly'].value_counts())
```

```python
# Visualize the anomalies on a 2D scatter plot
plt.figure(figsize=(12, 8))

# Plot normal points
plt.scatter(data[data['lof_anomaly'] == 1]['trip_distance'],
            data[data['lof_anomaly'] == 1]['fare_amount'],
            c='blue', label='Normal Points', alpha=0.6, s=20)

# Plot anomalous points
plt.scatter(data[data['lof_anomaly'] == -1]['trip_distance'],
            data[data['lof_anomaly'] == -1]['fare_amount'],
            c='red', label='Anomalies', alpha=0.8, s=50, marker='X')

plt.title('Anomaly Detection using Local Outlier Factor (NYC Taxi Fare Data)')
plt.xlabel('Trip Distance')
plt.ylabel('Fare Amount')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

#### **2. Anomaly Detection in Streaming Data**

For real-time anomaly detection in streaming data, a combination of approaches would be most robust. The choice depends on the nature of anomalies expected and computational constraints.

*   **Primary Choice: Isolation Forest (or a streaming variant)**
    *   **Why:** Isolation Forest is generally very efficient and scalable, making it suitable for high-velocity streaming data. It doesn't require distance calculations between all points, making it faster than LOF or DBSCAN for large datasets. It also works well in higher dimensions, which is likely when including various weather features and derived time features.
    *   **How:** Instead of training on the entire historical dataset, I would implement an online or incremental Isolation Forest. This involves periodically retraining the model on a sliding window of recent data or using algorithms designed for streaming data (e.g., `iForestASD`). Anomalies would be flagged when a new incoming data point's anomaly score exceeds a predefined threshold.

*   **Secondary/Complementary Choices (for specific use cases or further investigation):**
    *   **Local Outlier Factor (LOF):** While more computationally intensive, LOF is excellent at detecting local anomalies where density varies across the data space. It could be used as a secondary check or for offline analysis of flagged periods to confirm anomalies identified by Isolation Forest, especially if anomalies might appear in clusters of varying densities (e.g., a sudden drop in a region with historically lower usage). For streaming, it would likely be applied to recent batches of data.
    *   **DBSCAN:** DBSCAN is good for identifying clusters and noise, but its sensitivity to parameters (`eps`, `min_samples`) and struggle with varying densities make it less ideal for diverse streaming power grid data without significant parameter tuning per sub-region or condition. It could be used for identifying anomalous *clusters* of usage patterns rather than individual point anomalies.

*   **Real-time Implementation:**
    *   **Sliding Window:** Data points would be processed in mini-batches or a sliding window. The anomaly detection model is trained on data within this window or incrementally updated.
    *   **Thresholding:** A dynamic threshold for anomaly scores (perhaps adjusted based on historical false positive/negative rates) would trigger alerts.
    *   **Alerting System:** Integration with an alerting system to notify operators immediately about spikes or drops.

#### **3. Time Series Model for Short-Term Forecasting**

Given the 15-minute data, seasonality, trends, and the presence of exogenous variables (weather, region), **SARIMAX** would be the most appropriate choice for short-term forecasting.

*   **Why SARIMAX?**
    *   **Seasonality:** The 15-minute data will exhibit strong daily (96 observations/day) and weekly (96 * 7 observations/week) seasonality, and potentially monthly/yearly patterns. SARIMAX explicitly handles both non-seasonal and seasonal components.
    *   **Exogenous Variables:** The inclusion of weather conditions (temperature, humidity, etc.) and potentially region (as a categorical exogenous variable or using separate models) is crucial for accurate demand forecasting. SARIMAX (the 'X' part) allows for the incorporation of these external predictors.
    *   **Differencing (`I` component):** Real-time energy demand data is almost certainly non-stationary, exhibiting trends and seasonality. The 'I' (Integrated) component handles this by differencing the series to make it stationary.
    *   **AR and MA components:** These capture the autocorrelations and moving average of the series, ensuring the model learns from past values and past forecast errors.

*   **Implementation:**
    *   **Model per Region/Segment:** Depending on the granularity required and computational resources, it might be beneficial to train separate SARIMAX models for different regions or customer segments, as their demand patterns and weather sensitivities might differ.
    *   **Rolling Forecast Origin:** The model would be re-estimated or re-fitted periodically (e.g., daily, weekly) to incorporate the latest data and adapt to subtle shifts in patterns. For real-time 15-minute forecasts, a fixed model would be used, but its parameters would be updated frequently.
    *   **Forecast Horizon:** For short-term needs (e.g., next 1-4 hours), the SARIMAX model would generate forecasts for the next 4-16 time steps (15-minute intervals).

#### **4. Validation and Monitoring Performance Over Time**

Continuous validation and monitoring are critical in a real-time system.

*   **Anomaly Detection:**
    *   **Metrics:** Precision, Recall, F1-score, and False Positive Rate (FPR). Since anomalies are rare, F1-score is often more informative than accuracy.
    *   **Human-in-the-Loop:** A feedback loop where operators can confirm or reject flagged anomalies is essential. This ground truth data would be used to retrain and fine-tune the anomaly detection model periodically.
    *   **Concept Drift Detection:** Monitor the distribution of normal data and anomaly scores. If the characteristics of 'normal' change (e.g., due to new grid infrastructure, smart meter rollout), the model needs recalibration.

*   **Forecasting:**
    *   **Metrics:** Root Mean Squared Error (RMSE), Mean Absolute Error (MAE), Mean Absolute Percentage Error (MAPE). These would be tracked continuously for each forecast horizon.
    *   **Backtesting:** Regularly run the model on historical data to simulate its performance over past periods, especially during unusual events.
    *   **Residual Analysis:** Continuously monitor the residuals (actual - forecast). Non-random patterns in residuals (e.g., increasing bias, seasonality) indicate model degradation or a need for retraining.
    *   **Model Retraining:** Implement an automated pipeline to retrain the SARIMAX model at regular intervals (e.g., daily, weekly) or when performance metrics degrade significantly.
    *   **Comparison with Baselines:** Always compare the SARIMAX model's performance against simpler baseline models (e.g., naive forecasts, seasonal naive) to ensure added complexity is justified.
